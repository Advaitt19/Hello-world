```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv("star_classification.csv")
```

```python
data.head()
```

```python
data = data.drop(columns=["obj_ID","spec_obj_ID","run_ID", "rerun_ID", "cam_col", "field_ID", "plate", "MJD", "fiber_ID"])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
X = data.drop("class", axis = 1 )
y = data["class"]
y = LabelEncoder().fit_transform(y)
data.dropna(inplace=True)
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y, stratify = y,test_size = 0.2 , random_state = 42)
```

```python
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
Gt = GradientBoostingClassifier(n_estimators=100, random_state=42)
model_gt = Gt.fit(X_train,y_train)
```

```python
y_pred_1 = model_gt.predict(X_test)
```

```python
accuracy_base = accuracy_score(y_test,y_pred_1)
print(accuracy_base)
```

```python
Gt = GradientBoostingClassifier(n_estimators=100, random_state=42)
model_gt_s = Gt.fit(X_train_s,y_train)
```

```python
y_pred_1_2 = model_gt_s.predict(X_test_s)
```

```python
accuracy_base_s = accuracy_score(y_test,y_pred_1_2)
print(accuracy_base_s)
```

```python
GBt_Hpt = GradientBoostingClassifier()
```

```python
param_grid = {"n_estimators": [100],
              "learning_rate": [0.01],
              "max_depth": [3],
              }
```

```python
grid = GridSearchCV(GBt_Hpt, param_grid, cv=3, scoring="accuracy", n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
y_pred_2 = grid.predict(X_test)
```

```python
accuracy_Main = accuracy_score(y_pred_2,y_test)
print(accuracy_Main)
```

```python
Gt = GradientBoostingClassifier(random_state=42)
model_gt_1 = Gt.fit(X_train,y_train)
y_pred_zero = model_gt_1.predict(X_test)
accuracy_base_zero = accuracy_score(y_test,y_pred_zero)
print(accuracy_base_zero)
```
